## Analyse dataset 2

```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_excel("dataset 2 - aide-exceptionnelle-cpsti-covid-19.xlsx",sheet_name="aide-exceptionnelle-cpsti-covid")
df.head()
```

```python
labels=['Dispositif','Code ancienne région','Ancienne région','Code département','Code NA21','Geo Point']
labels+=['Montant en millions d\'euros']
df=df.drop(labels, axis=1)
```

```python
df_reg=df.groupby(['Code région']).sum()
df_reg['Montant total de l\'aide']=df_reg['Montant total de l\'aide']//1000
df_reg=df_reg.reset_index()
df_reg=df_reg.drop(index=17)
df_reg.head(20)
```

```python
import folium
regions_map = r'regions.geojson'
france_map = folium.Map(location=[45.7833, 3.0833], zoom_start=5, max_bounds=True)
folium.Choropleth(
    geo_data=regions_map,
    data=df_reg,
    columns=['Code région','Montant total de l\'aide'],
    key_on='feature.properties.code',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Total aide en millieurs d euros'
).add_to(france_map)

# display map
france_map
#france_map.save('map.html')
```
